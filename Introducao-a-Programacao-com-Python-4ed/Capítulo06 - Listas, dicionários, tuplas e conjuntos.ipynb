## 6.7 Uso de listas como pilhas

```python
# Programa 6.8: Pilha de pratos
prato = 5
pilha = list(range(1, prato + 1))

while True:
    print(f"\nExistem {len(pilha)} pratos na pilha")
    print(f"Pilha atual: {pilha}")
    print("Digite E para empilhar um novo prato,")
    print("ou D para desempilhar. S para sair.")
    
    operação = input("Operação (E, D ou S):").upper()
    
    if operação == "D":
        if len(pilha) > 0:
            lavado = pilha.pop(-1)
            print(f"Prato {lavado} lavado")
        else:
            print("Pilha vazia! Nada para lavar.")
            
    elif operação == "E":
        prato += 1  # Novo prato
        pilha.append(prato)
        
    elif operação == "S":
        break
        
    else:
        print("Operação inválida! Digite apenas E, D ou S!")
```

Exercício 6.7 - Faça um programa que leia uma expressão com parênteses. Usando pilhas, verifique se os parênteses foram abertos e fechados na ordem correta. Exemplo:


(()) OK   
()()(()()) OK   
()) Erro   


Você pode adicionar elementos à pilha sempre que encontrar abre parênteses
e desempilhá-la a cada fecha parênteses. Ao desempilhar, verifique se o topo
da pilha é um abre parênteses. Se a expressão estiver correta, sua pilha estará
vazia no final.

```python
expression = input("Digite a expressão com parênteses: ")
pilha = []
error = False

for caractere in expression:
    if caractere == "(":
        pilha.append("(") # Empilha
    elif caractere == ")":
        if len(pilha) > 0:
            pilha.pop(-1) # Desempilha o último
        else:
            error = True # Teve ')' sem o '(' correspondente. OU seja, está dando pop sem ter elementos para remover
if not error and len(pilha) == 0:
    print("OK")
else:
    print("Erro")
```

Exercício 6.11 - Modifique o Programa 6.6 usando for. Explique por que nem todos os while podem ser transformados em for.

Nem todos os while podem ser convertidos em for pois, em alguns algoritmos, o objetivo é processar um (ou vários) elementos de uma lista em um índice específico (fatiamento), de forma não sequencial, coisa que o laço for não faz já que este começa necessariamente do primeiro elemento até o último. O laço for, por sua vez, processa elemento a elemento retornando seu valor (e não seu índice na lista). Por outro lado, o laço while é usado quando não sabemos quantas vezes vamos repertir aquele bloco ou quando queremos processar um objeto de forma não sequencial.

```python
# Programa 6.6
T = []
while True:
    n = int(input("Digite um número (0 para sair): "))
    if n == 0:
        break
    T.append(n)

x = 0
while x < len(T):
    print(T[x])
    x += 1
```

## 6.11 - Enumerate. 
A função enunerate gera uma tupla (par de valores) em que o primeiro valor é o índice e o segundo é o elemento da lista sendo enumerada.

```python
T = [5, 9, 13]
x = 0

for e in T:
    print(f"[{x}] : {e}")
    x += 1

#Mesma coisa, só que com enumerate
for x, e in enumerate(T):
    print(f"[{x}] : {e}")

#Demonstrando que enumerate() gera uma tupla, que pode ser desempacotada

for z in enumerate(T):
    x, e = z
    print(f"[{x}] : {e}")
    print("Tupla: ", z)
```

## 6.13 - Aplicações

```python
# Programa 6.13: Controle da utilização de salas de um cinema
lugares_vagos = [10, 2, 1, 3, 0]

while True:
    sala = int(input("Sala (0 sai): "))
    if sala == 0:
        print("Fim")
        break
    if sala > len(lugares_vagos) or sala < 1:
        print("Sala inválida")
    elif lugares_vagos[sala - 1] == 0:
        print("Desculpe, sala lotada!")
    else:
        lugares = int(input(f"Quantos lugares você deseja ({lugares_vagos[sala - 1]} vagos): "))
        if lugares > lugares_vagos[sala - 1]:
            print("Esse número de lugares não está disponível.")
        elif lugares < 0:
            print("Número inválido")
        else:
            lugares_vagos[sala - 1] -= lugares
            print(f"{lugares} lugares vendidos")

print("\nUtilização das salas")
for sala, vagos in enumerate(lugares_vagos):
    print(f"Sala {sala + 1} - {vagos} lugar(es) vazio(s)")
```

## 6.16 - Ordenação

buble-sort

```python
# Programa 6.20

L = [7, 4, 3, 12, 8]
fim = 5
while fim > 1:
    trocou = False
    x = 0
    while x < (fim - 1):
        if L[x] > L[x + 1]:
            trocou = True
            temp = L[x]
            L[x] = L[x + 1]
            L[x + 1] = temp
        x += 1
    if not trocou:
        break
    fim -= 1

for e in L:
    print(e)
    
print(L)
```

O método .sort() ordena alterando a ordem da lista na variável original. A função sorted() apenas retorna a lista ordenada, porém, não modifica a ordem na variável original.

```python
L = [7, 4, 3, 12, 8]
print("Lista original: ", L)
print("Função sorted():", sorted(L), "sem alterar a lista original: ", L)
print("Perceba que a função sorted() não alterou a lista")
print("\nJá o método .sort() alterou a lista original:")
L.sort()
print(L)
```

## 6.17 - Dicionários

Estrutura de dados similar às listas, porém, os dicionários guardam um par de chave:valor (sempre nessa ordem). A vantagem dos dicionários reside em ter sempre relacionado uma chave ao seu valor correspondente. 

O dict pode ser acessado através das suas chaves. dicionario["chave"] retornará o valor atrelado à "chave". Portanto, diferentemente das listas, as chaves são os índices. Se, ao tentar adicionar um valor à uma chave já existente, o dict será alterado naquele valor específico. Caso seja uma nova chave, ela será adicionada ao dicionário.

Ao tentar acessar uma chave inexistente, o interpretador retornará KeyError

```python
tabela = {
    "Alface": 0.45,
    "Batata": 1.20,
    "Tomate": 2.30,
    "Feijão": 1.50
}

print(tabela)
print(tabela["Tomate"])
```

```python
tabela["Tomate"] = 3.00 #chave existente -> atualização do seu valor
print(tabela["Tomate"])
tabela["Maçã"] = 5.50 #Nova chave
print(tabela)
print(tabela["Manga"]) #KeyError chave inexistente
```

```python
#Verificando presença de uma chave em um dict

print("Feijão" in tabela)
print("Leite" in tabela)

#Lista com chaves e valores presentes
print(tabela.keys())
print(tabela.values())
```

```python
#Programa 6.21 - Obtenção do preço com um dict

tabela = {
    "Alface": 0.45,
    "Batata": 1.20,
    "Tomate": 2.30,
    "Feijão": 1.50
}

while True:
    produto = input("Digite o nome do produto (fim para terminar): ")
    if produto == "fim":
        break
    if produto in tabela:
        print(f"Preço de {produto} é R$ {tabela[produto]:5.2f}")
    else:
        print("Produto não encontrado!")
```

Dicionários podem, também, conter listas dentro deles (ou até outros dicionários)

```python
# Programa 6.22: Exemplo de dicionário com estoque e operações de venda
estoque = {
    "tomate": [1000, 2.30],
    "alface": [500, 0.45],
    "batata": [2001, 1.20],
    "feijão": [100, 1.50]
}

venda = [["tomate", 5], ["batata", 10], ["alface", 5]]
total = 0

print("Vendas:\n")

for operação in venda:
    produto, quantidade = operação
    preço = estoque[produto][1]
    custo = preço * quantidade
    print(f"{produto:12s}: {quantidade:3d} x {preço:6.2f} = {custo:6.2f}")
    estoque[produto][0] -= quantidade
    total += custo

print(f"Custo total: {total:21.2f}\n")
print("Estoque:\n")

for chave, dados in estoque.items(): 
    print("Descrição: ", chave)
    print("Quantidade: ", dados[0])
    print(f"Preço: {dados[1]:6.2f}\n")
```

### 6.19 Dicionários com valor padrão

O método .get() retorna o valor de uma chave. Caso ela não exista, o .get(parâmetro1, parâmetro2) retorna o segundo parâmetro. Caso ele não seja especificado, .get() retorna None.

```python
# Programa 6.20 usando método .get()

frase = input("Digite uma frase: ")
base = {}

for caractere in frase:
    base[caractere] = base.get(caractere, 0) + 1
print(base)
```

## 6.20 - Tuplas
Strings separadas por vírgulas são tuplas pois são imutáveis. Opcionalmente, podemos delimitar tuplas por parênteses, mas sempre usando vírgulas para separar os elementos. t=(1) é um inteiro. t=(1,) é uma tupla com um elemento.

```python
palavra = "amor"
a = "teste", "teste2"
b = "a", "b", "c"
c = ("a", "b")

print(type(palavra))
print(type(a))
print(type(b))
print(type(c))
```

A tupla pode conter outros objetos e estes podem ser mutáveis. Objetos originalmente multáveis permanecem com essa propriedade apesar de estarem dentro de uma tupla.

```python
tupla = ("a", "b", ["c", "d", "e"])

print(len(tupla))
print(tupla[2])

tupla[2].append("f")
print(tupla)
```

### Empacotamento e desempacotamento com *

O * indica "todos em..." Ou seja, a variável sem * receberá apenas 1 valor enquanto que a marcada com * receberá todo o restante presente na lista.

```python
*a, b = [1, 2, 3, 4, 5] #último em b e todos os primeiros em a
print(b)
print(a)

a, *b = [1, 2, 3, 4, 5] #primeiro em a e todos os demais em b
print(a)
print(b)

```

Assim, se tivermos *a, b, c, d dizemos que todos os valores serão alocado em a exceto os unitários colocados em b, c, d (neste caso, os 3 últimos valores). Outro exemplo:

```python
a, *b, c = [1, 2, 3, 4, 5] #primeiro em a, último em c e todos os demais em b
print(a)
print(b)
print(c)
```

## 6.21 Conjuntos (set)

São estruturas de dados que não admitem repetição e não são ordenados, não suportando acesso por índice (posição).

```python
a = set()
a.add(1)
a.add(3)
a.add(5)
a.add(-1)
print(a)

b = [6,7,8]
b= set(b)
print(b)

print(3 in a)
print(3 in b)
```

União ( | ) unir os dois conjuntos

```python
a = {0, 1, 2, 3, -1}
b = {2, 3}
c = set([4,5,6])

print(a | b)
print(a | c)
```

Interseção ( & ) elementos que estão presentes em ambos os conjuntos

```python
A = {1,2,3,6}
B = {2,4,6}

print(A & B)
```

Diferença ( - ) elementos de A não contudos em B

```python
A = {1,2,3,6}
B = {2,4,5,6}

print(A - B)
print(B - A)
```

Diferença simétrica ( ^ ) elementos que são exclusivos de cada conjunto isoladamente. Elementos não comuns aos dois conjuntos

```python
A = {1,2,3,6}
B = {2,4,5,6}

print(A ^ B)
```

```python
#Outras operações possíveis em conjuntos

A = {x for x in range(0, 10, 2)}
print(A)
B = {2, 4}

# verificar se A contém B >
print(A > B) # Todos os elementos de B estão contidos em A
print(B > A) # Nem todo elemento de A está contido em B
```

```python
# Quando os conjuntos têm os mesmos elementos ou são iguais, o resultado de > é False
print(A > A)
print(B > {2,4})

# Um conjunto só é maior ou igual a outro quanto contém esse outro
print(A >= A)
print(B >= B)
```

```python
#O inverso é valido para <

print(A < B) # A está contido em B?
print(B < A) # B está contido em A?
```

```python
# Da mesma forma, use == ou != para verificar se os conjuntos possuem os mesmos elementos 
# ou se tem elementos diferentes

print(A == B)
print(B == B)
print(B != A)
```

- os valores comuns às duas listas
- os valores que só existem na primeira
- os valores que existem apenas na segunda
- uma lista com os elementos não repetidos das duas listas
- a primeira lista sem os elementos repetidos na segunda

```python
lista1 = [x for x in range(1, 10)]
lista2 = [x for x in range(6, 16)]
set1 = set(lista1)
set2 = set(lista2)
print(f"Lista 1: {set1}")
print(f"Lista 2: {set2}")
print(f"Valores comuns às duas listas: {set1 & set2}")
print(f"Valores que só existem na lista 1: {set1 - set2}")
print(f"Valores que só existem na lista 2: {set2 - set1}")
print(f"Uma lista com elementos não repetidos das duas listas: {list(set1 | set2)}")


```

Exercício 6.22 - Escreva um programa que compare duas listas. Considere a pri­meira lista como a versão inicial e a segunda como a versão após alterações. Utilizando operações com conjuntos, seu programa deverá imprimir a lista de modificações entre essas duas versões, listando:

- os elementos que não mudaram
- os novos elementos
- os elementos que foram removidos

```python
inicial = [x for x in range(5,9)]
final = [x for x in range(7,11)]
#Conversaão para operações com set

set_inicial = set(inicial)
set_final = set(final)

print(f"Lista inicial {inicial}")
print(f"Lista final {final}")
print(f"Não mudaram: {set_inicial & set_final}")
print(f"Novos elementos:{set_final - set_inicial}")
print(f"Elementos removidos:{set_inicial - set_final}")
```
