Exercício 4.2 - Escreva um programa que pergunte a velocidade do carro de um usuário. Caso ultrapasse 80 km/h, exiba uma mensagem dizendo que o usuá­rio foi multado. Nesse caso, exiba o valor da multa, cobrando R$ 5 por km
acima de 80 km/h.

```python
velocidade = float(input("Qual a velocidade do carro? "))

if velocidade > 80:
  multa = (velocidade - 80) * 5
  print(f"Você foi multado. O valor da multa é de R${multa:5.2f}")
else:
  print("Parabéns! Você não recebeu multa!")
```

```python
salario = float(input("Qual o salário do funcionário? "))

if salario > 1_250:
  salario = salario + (0.1 * salario)
  print(f"O novo salário foi acrescido em 10% ficando, agora, {salario}")
if salario <= 1_250:
  salario = salario + (0.15 * salario)
  print(f"O novo salário foi acrescido em 15% ficando, agora, {salario}")
```

Exercício 4.10 - Escreva um programa que leia dois números e que pergunte qual operação você deseja realizar. Você deve poder calcular soma (+), subtração (-),
multiplicação (*) e divisão (/). Exiba o resultado da operação solicitada.

```python
a = float(input("Digite o primeiro número: "))
op = input("Digite a operação desejada: soma (+), subtração (-), multiplicação (*) e divisão (/)")
b = float(input("Digite o segundo número: "))


if op == "soma" or op == "+":
  print(f"A soma de {a} + {b} é {a + b}")

elif op == "subtração" or op == "-":
  print(f"A subtração de {a} - {b} é {a - b}")

elif op == "multiplicação" or op == "*":
  print(f"A multiplicação de {a} por {b} é {a * b}")

elif op == "divisão" or op == "/":
  print(f"A divisão de {a} por {b} é {a / b}")

else:
  print(f"Operação {op} inválida!")
```

Exercício 4.11 - Escreva um programa para aprovar o empréstimo bancário para compra de uma casa. O programa deve perguntar o valor da casa a comprar, o salário e a quantidade de anos a pagar. O valor da prestação mensal não pode ser superior a 30% do salário. Calcule o valor da prestação como sendo o valor da casa a comprar dividido pelo número de meses a pagar

```python
preco_casa = float(input("Qual o preço da casa? "))
salario = float(input("Qual seu salário? "))
tempo_anos = int(input("Quantos anos a pagar? "))

prestacoes = preco_casa / (tempo_anos * 12)

if prestacoes > 0.3 * salario:
  print(f"Desculpe, você não tem condições financeiras para pagar esta casa no tempo proposto. A prestação de R${prestacoes} excedem 30% do seu salário.")

else:
  print(f"Parabéns. Crédito aprovado! Suas prestações serão de R${prestacoes:5.2f} ao mês.")
```
