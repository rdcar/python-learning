Exercício 5.13 Escreva um programa que pergunte o valor inicial de uma dívida e o juro mensal. Pergunte também o valor mensal que será pago. Imprima o número de meses para que a dívida seja paga, o total pago e o total de juros pago.

````python
divida_inicial = float(input("Valor inicial da dívida (R$): "))
taxa_juros = float(input("Juro mensal (%): ")) / 100
pagamento_mensal = float(input("Valor mensal a ser pago (R$): "))

divida_restante = divida_inicial
total_pago = 0
total_juros = 0
meses = 0

# Verificação inicial (para evitar loop infinito onde o juro é maior que o pagamento)
juro_primeiro_mes = divida_inicial * taxa_juros

if pagamento_mensal <= juro_primeiro_mes:
    print("\nERRO: O pagamento mensal não cobre sequer o juro do primeiro mês. A dívida é impagável e crescerá indefinidamente.")
else:
    print("\n--- Simulação Iniciada ---")

while divida_restante > 0:
    juros_do_mes = divida_restante * taxa_juros
    total_juros += juros_do_mes
    divida_restante += juros_do_mes
    pagamento_realizado = min(pagamento_mensal, divida_restante) # para que, no último mês, o valor pago não ultrapasse a dívida
    divida_restante -= pagamento_realizado
    total_pago += pagamento_realizado
    meses += 1

    print(f"Mês {meses:2d}: Juros: R$ {juros_do_mes:7.2f} | Pago: R$ {pagamento_realizado:7.2f} | Resta: R$ {divida_restante:7.2f}")

print("\n--- Resultado Final ---")
print(f"Número de meses para que a dívida seja paga: {meses}")
print(f"Total pago: R$ {total_pago:10.2f}")
print(f"Total de juros pago: R$ {total_juros:10.2f}")
````

Exercício 5.23 Escreva um programa que leia um número e verifique se é ou não um número primo. Para fazer essa verificação, calcule o resto da divisão do número por 2 e depois por todos os números ímpares até o número lido. Se
o resto de uma dessas divisões for igual a zero, o número não é primo. Observe que 0 e 1 não são primos e que 2 é o único número primo que é par.

````python
while True:
    numero = int(input("Entre com um número natural (0 para sair): "))
    
    if numero == 0:
        print("Encerrando programa...")
        break
    
    if numero < 2:
        print(f"{numero} não é primo (0 e 1 NÃO são primos).")
    elif numero == 2:
        print("2 é PRIMO!")
    elif numero % 2 == 0:
        print(f"{numero} NÃO É PRIMO! (É par e maior que 2)")
    else:
        # Aqui testamos os ímpares a partir de 3
        divisor = 3
        e_primo = True # Assumimos que é primo até provar o contrário
        
        while divisor < numero:
            if numero % divisor == 0:
                e_primo = False # Encontramos um divisor, então não é primo
                break
            divisor = divisor + 2 # Pula de 2 em 2 para testar apenas ímpares
            
        if e_primo:
            print(f"{numero} é PRIMO!")
        else:
            print(f"{numero} NÃO É PRIMO!")
````

Dica de Ouro (Performance): Para números muito grandes, você não precisa testar até o próprio número. Basta testar até a raiz quadrada do número. Se nenhum número até a raiz quadrada o dividir, nenhum outro acima dela dividirá.

## Explicação

Vamos usar uma analogia para separar as responsabilidades:

Imagine uma **Fábrica de Parafusos** onde queremos separar apenas os parafusos perfeitos (Primos).

---

### 1. O Laço Externo (`numero_atual += 2`)

**Função:** É a "Esteira de Produção".
Ele é o responsável por trazer novos candidatos para serem testados.

* Ele diz: "Agora vamos testar o número 3".
* Depois que o teste termina, ele diz: "Ok, agora traga o 5".
* Ele incrementa de 2 em 2 (`numero_atual += 2`) porque, como já testamos o 2 lá no início, não faz sentido a esteira trazer números pares (4, 6, 8...), pois já sabemos que eles não são primos.
* **Velocidade:** Ele muda devagar. Ele espera todo o teste do laço interno acabar para só então mudar para o próximo número.

---

### 2. O Laço Interno (`divisor += 2`)

**Função:** É o "Inspetor de Qualidade".
Ele não muda o número que está na esteira; ele tenta **quebrar** o número que o laço externo forneceu.

* Se o laço externo trouxe o número **15**, o laço interno começa a trabalhar:
* Testa `15 % 3`. Deu zero? "Opa, achei um divisor! Não é primo".


* Se o laço externo trouxe o número **11**, o laço interno:
* Testa `11 % 3`... não deu zero.
* **Incrementa o divisor:** Testa `11 % 5`... não deu zero.
* O `divisor += 2` serve para o inspetor não ficar testando o mesmo número toda hora. Ele precisa tentar o próximo divisor ímpar possível.


* **Velocidade:** Ele é muito rápido. Ele gira várias vezes para **cada um** dos giros do laço externo.

---

### Visualizando o movimento:

Se o código quer encontrar os primeiros primos, o movimento das variáveis seria assim:

1. **Externo** traz o **9**.
* **Interno** testa divisor **3**: `9 % 3 == 0`? **SIM!**
* O interno para (`break`) e diz: "9 não presta".


2. **Externo** pula para o **11** (`numero_atual += 2`).
* **Interno** testa divisor **3**: `11 % 3`? Não.
* **Interno** aumenta para **5** (`divisor += 2`): `11 % 5`? Não.
* O interno termina porque o divisor ficou alto demais. O fiscal anota: "11 é primo!".


3. **Externo** pula para o **13**... e o ciclo recomeça.

### Resumo do porquê incrementar ambos:

* Você incrementa o **`divisor`** para testar novas possibilidades de divisão para o **mesmo** número.
* Você incrementa o **`numero_atual`** para mudar o foco e começar a testar um **novo** número do zero.

Se você esquecesse de incrementar o `divisor`, o programa ficaria preso testando `15 % 3` para sempre (**loop infinito**). Se esquecesse o `numero_atual`, você testaria o número 3 para sempre.

Exercício 5.25 Escreva um programa que calcule a raiz quadrada de um número. Utilize o método de Newton para obter um resultado aproximado. Sendo n o número a obter a raiz quadrada, considere a base b=2. Calcule p usando a
fórmula p=(b+(n/b))/2. Agora, calcule o quadrado de p. A cada passo, faça b=p e recalcule p usando a fórmula apresentada. Pare quando a diferença absoluta entre n e o quadrado de p for menor que 0,0001. 

````python
n = float(input("Entre com o número a obter a raiz quadrada:"))
b = 2
p = (b + (n/b)) / 2
p2 = p ** 2
i = 1 # Contador

while abs(p2 - n) >= 0.0001:
    b = p
    p = (b + (n/b)) / 2
    p2 = p ** 2
    # Eu quis botar contador pra ver quantas vezes o laço era executado até convergir.
    print(f"Essa é a {i} iteração") 
    i += 1
    
print(f"A raiz quadrada de {n} pela aproximação de Newton é {p:.4f} .")
````

Exercício 5.27 Escreva um programa que verifique se um número é palíndromo.
Um número é palíndromo se continua o mesmo caso seus dígitos sejam invertidos. Exemplos: 454,10501

````python
n = int(input("Entre com um número para verificação:"))
original = n
invertido = 0

#Basicamente rescreve os digitos ao contrario usando resto e inteiro da divisão enquanto n > 0
while n > 0:
    digito = n % 10 #Extrai o útimo digito
    invertido = invertido * 10 + digito #Desloca o invertido uma casa para esquerda
    n //= 10 #Remove o digito usado

if original == invertido:
    print(f"É palíndromo! {original} = {invertido}")
else:
    print(f"Não é palíndromo! {original} diferente de {invertido}")
````
