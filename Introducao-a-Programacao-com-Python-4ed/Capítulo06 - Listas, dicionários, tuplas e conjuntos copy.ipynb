Exercício 6.3 -Faça um programa que percorra duas listas e gere uma terceira sem elementos repetidos.

```python
A = [True, "a","b", "c", 1, 2, 3]
B = [2, 3, 4, 5, "a", "c", 8]
C = []

for i in A:
    if i not in C:
        C.append(i)

for j in B:
    if j not in C:
        C.append(j)

print(f"Lista C (sem repetidos): {C}")
```

6.6 Usando listas como filas

```python
# Programa 6.7: Simulação de uma fila de banco
último = 10
fila = list(range(1, último + 1))
while True:
    print(f"\nExistem {len(fila)} clientes na fila")
    print(f"Fila atual: {fila}")
    print("Digite F para adicionar um cliente ao fim da fila,")
    print("ou A para realizar o atendimento. S para sair.")
    operação = input("Operação (F, A ou S):")
    if operação == "A":
        if len(fila) > 0:
            atendido = fila.pop(0)
            print(f"Cliente {atendido} atendido")
        else:
            print("Fila vazia! Ninguém para atender.")
    elif operação == "F":
        último += 1
        # Incrementa o ticket do novo cliente
        fila.append(último)
    elif operação == "S":
        break
    else:
        print("Operação inválida! Digite apenas F, A ou S!")
```

Exercício 6.5 - Altere o Programa 6.7 de forma a poder trabalhar com vários comandos digitados de uma só vez. Atualmente, apenas um comando pode ser inserido por vez. Altere-o de forma a considerar operação como uma string.


Exemplo: FFFAAAS significaria três chegadas de novos clientes, três atendi­mentos e, finalmente, a saída do programa.

```python
# Programa 6.7: Simulação de uma fila de banco
último = 10
fila = list(range(1, último + 1))

while True:
    print(f"\nExistem {len(fila)} clientes na fila")
    print(f"Fila atual: {fila}")
    print("Digite F para adicionar um cliente ao fim da fila,")
    print("ou A para realizar o atendimento. S para sair.")
    operação = input("Operação (F, A ou S):")
    sair = False
    for opção in operação:
        if opção == "A":
            if len(fila) > 0:
                atendido = fila.pop(0)
                print(f"Cliente {atendido} atendido")
            else:
                print("Fila vazia! Ninguém para atender.")
        elif opção == "F":
            último += 1
            # Incrementa o ticket do novo cliente
            fila.append(último)
            print(f"Cliente {último} adicionado")
        elif opção == "S":
            sair = True
        else:
            print("Operação inválida! Digite apenas F, A ou S!")
    if sair:
        print("Saindo...")
        break
```

Exercício 6.6 - Modifique o programa para trabalhar com duas filas. Para facili­tar seu trabalho, considere o comando A para atendimento da fila 1; e B, para atendimento da fila 2. O mesmo para a chegada de clientes: F para fila 1; e G, para fila 2.

```python
# Programa 6.7: Simulação de uma fila de banco
último = 10
fila1= list(range(1, último + 1))
fila2= list(range(1, último + 1))

while True:
    print(f"\nExistem {len(fila)} clientes na fila")
    print(f"Fila atual: {fila}")
    print("Digite F para adicionar um cliente ao fim da fila,")
    print("ou A para realizar o atendimento. S para sair.")
    operação = input("Operação (F, A ou S):")
    sair = False
    for opção in operação:
        if opção == "A":
            if len(fila1) > 0:
                atendido = fila1.pop(0)
                print(f"Cliente {atendido} atendido na Fila 1")
            else:
                print("Fila 1 vazia! Ninguém para atender.")
        elif opção == "B":
            if len(fila2) > 0:
                atendido = fila2.pop(0)
                print(f"Cliente {atendido} atendido na Fila 2")
            else:
                print("Fila 2 vazia! Ninguém para atender.")
        elif opção == "F":
            último += 1
            # Incrementa o ticket do novo cliente
            fila1.append(último)
            print(f"Cliente {último} adicionado à Fila 1")
        elif opção == "G":
            último += 1
            # Incrementa o ticket do novo cliente
            fila2.append(último)
            print(f"Cliente {último} adicionado à Fila 2")

        elif opção == "S":
            sair = True
        else:
            print("Operação inválida! Digite apenas F, A, B, G ou S!")
    if sair:
        print("Saindo...")
        break
```
