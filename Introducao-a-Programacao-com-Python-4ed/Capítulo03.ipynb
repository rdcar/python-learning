Os operadores lógicos podem ser combinados em expressões lógicas mais com­plexas. Quando uma expressão tiver mais de um operador lógico, avalia-se o operador **not** (não) primeiro, seguido do operador **and** (e) e, finalmente, **or** (ou).

Vejamos a seguir a ordem de avaliação da expressão, em que a operação sendo avaliada é **destacada em negrito**; e o resultado, mostrado na linha seguinte.

---

True or False and <u>**not True**</u>  
True or <u>**False and False**</u>  
<u>**True or False**</u>  
True

---

Os operadores lógicos também podem ser usados em conjunto com os operadores relacionais (> < >= <=). Nesse caso, os operadores relacionais devem ser avaliados primeiro.

---

salário = 100, idade = 20

salário > 1000 and idade > 18   
**100 > 1000** and **20 > 18**   
False **and** True   
False

Exercício 3.9 Escreva um programa que leia a quantidade de dias, horas, minu­tos e segundos do usuário. Calcule o total em segundos.

```python
dias = int(input("Digite a quantidade de dias: "))
horas = int(input("Digite a quantidade de horas: "))
minutos = int(input("Digite a quantidade de minutos: "))
segundos = int(input("Digite a quantidade de segundos: "))

"""
Minutos para Segundos: 1 minuto = 60 segundos
Horas para Segundos: 1 hora = 60 minutos x 60 segundos/minuto = 3.600 segundos
Dias para Segundos: 1 dia = 24 horas X 3.600 segundos/hora = 86.400 segundos

PEP8 -> usar costantes com variáveis em MAIÚSCULO
"""

SEGUNDOS_POR_MINUTO = 60
SEGUNDOS_POR_HORA = 3_600
SEGUNDOS_POR_DIA = 86_400
total = segundos + (SEGUNDOS_POR_MINUTO * minutos) + (SEGUNDOS_POR_HORA * horas) + (SEGUNDOS_POR_DIA * dias)

print(f"O total, em segundos, é {total}.")
```
